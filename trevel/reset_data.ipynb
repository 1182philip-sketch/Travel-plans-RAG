# 데이터 초기화
chat_history.csv / users.csv / favorites.csv 를 헤더만 남기고 초기화합니다.

```python
import pandas as pd

files = {
    "chat_history.csv": ["user_id", "role", "content", "timestamp"],
    "users.csv":        ["user_id", "preferences", "created_at"],
    "favorites.csv":    ["user_id", "title", "content", "saved_at"],
}

for filename, columns in files.items():
    pd.DataFrame(columns=columns).to_csv(filename, index=False, encoding="utf-8-sig")
    print(f"✅ {filename} 초기화 완료")

print("\n모든 데이터 초기화 완료!")
```
